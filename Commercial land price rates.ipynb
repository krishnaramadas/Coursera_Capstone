```python
#droping the unnecessary column
q=h.drop('Residential rates (All numbers in Rs Sq/ft)', axis=1)
#droping the column names whihc are added as individual observations in the dataframe
q.drop([0,16,25,29,36,46,52,60,67,76,85,105,108,122,131,139,152,169],inplace=True) 
q.head()
```

```python
w=[] #a null list to save all the avg value of each zone's land price 
a=q.iloc[1:16] #slicing the df to each zone
w.append(a.mean()) #finding mean and adding the value to the null list
d=q.iloc[47:67]
w.append(d.mean())
g=q.iloc[68:85]
w.append(g.mean())
j=q.iloc[86:105]
w.append(j.mean())
m=q.iloc[106:131]
w.append(m.mean())
p=q.iloc[132:139]
w.append(p.mean())
t=q.iloc[140:176]
w.append(t.mean())
w
```

```python
#mapping the avg list values to the respective zones and making a new dataframe
p_data={"Central": w[0], "Western": w[1], "Northern": w[2], "North-Eastern": w[3], "Eastern": w[4], "South-Eastern": w[5], "Southern" : w[6], "Southern Suburbs" : w[5]}
price_data=pd.DataFrame.from_dict(p_data)
price_data
```

```python
#transposing the dataframe to make a new dataframe with zones and Price as column
df=price_data.stack().astype(float)
pdf=df.to_frame().reset_index()
price_df=pdf.rename(columns={"level_0": " ", "level_1": "Zone", 0:"Average Price per Sq/Ft"})
price_df
```

```python
#saving the dataframe as a csv file
price_df.to_csv("price_df.csv", index=False)
```
